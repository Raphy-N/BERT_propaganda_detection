```python
def parse_label_test(label_path: str, binary: str = None) -> list:
    # idx, type, start, end
    labels = []
    dictd = {}
    f= Path(label_path)
    if not f.exists():
        return labels
    for line in open(label_path):
        parts = line.strip().split('\t')
        #print (parts)
        if parts[0] not in dictd:
            dictd[parts[0]] = []
        dictd[parts[0]].append((int(parts[2]), int(parts[3]), 'Propaganda'))
    return dictd
```

```python
dict_pred = parse_label_test("exp/binary/binary_test/pred.test", binary=True)
#id_check, pred_labels = parse_label_test("exp/binary/binary_test/pred.test", binary=True)
```

```python
dict_pred = dict(sorted(dict_pred.items()))
```

```python
dict_gold = dict(zip(ids, labels))
dict_gold = dict(sorted(dict_gold.items()))
```

```python
dict_text = dict(zip(ids, texts))
dict_text = dict(sorted(dict_text.items()))
```

```python
# prop. char count / total char count 
def get_measure(ids, texts, labels):
    dictd = {}
    for idx, text, label in zip(ids, texts, labels):
        char_article = 0 
        for prop in label:
            count = prop[1] - prop[0]
            char_article += count
        total_char = text[-1].idx
        dictd[idx] = char_article/total_char
    return dictd
```

```python
ids = list(dict_text.keys())
texts = [dict_text[k] for k in dict_text]
labels= [dict_gold[k] for k in dict_gold]
pred_labels = [dict_pred[k] for k in dict_pred]
```

```python
dictpred = get_measure(ids, texts, pred_labels)
dictgold = get_measure(ids, texts, labels)
```

```python
sortedpred = sorted(dictpred.items(), key=lambda x: x[1])
sortedgold = sorted(dictgold.items(), key=lambda x: x[1])

predict = [i[0] for i in sortedpred]
gold = [i[0] for i in sortedgold]
#print (sorted_prediction)
tau, p_value = stats.kendalltau(predict, gold)
print (tau, p_value)
```
